# **Jaringan Semantik Jenis Kuliner**

Dalam visualisasi jaringan semantik menggunakan python, langkah pertama yang harus dilakukan adalah mengimportkan library yang dibutuhkan. Terdapat dua library yang dibutuhkan, yaitu networkx dan pandas kemudian tetapkan sebagai nx dan pd

```python
import networkx as nx
import pandas as pd
```

Inputkan dataset untuk dilakukan pembacaan dengan menggunakan library pandas. Pada program ini, terdapat dua file dataset dengan ekstensi xlsx.
Assign kedua dataset pada variabel baru, variabel nodes_df digunakan untuk membaca dataset nodes yang berisi daftar nama item pada semantik, lalu variabel edges_df digunakan untuk memebaca dataset edges yang berisi daftar relation antar node. 

```python
nodes_df = pd.read_excel("/content/drive/MyDrive/Semantik Kuliner/nodes.xlsx")
edges_df = pd.read_excel("/content/drive/MyDrive/Semantik Kuliner/edges.xlsx")

# Tampilkan list nama node untuk melakukan search filtering pada bagian dibawah nanti
nodes_df
```

**Tampilan Jaringan Semantik Keseluruhan**

> Bagian ini digunakan untuk menampilkan hasil visualisasi jaringan semantik secara keseluruhan.
Prosedurnya adalah dengan menetapkan library networkx pada variabel baru, yaitu variabel G.
Kemudian lakukan perulangan untuk membaca baris data yang berisi nama node dan relasi antar node







```python
# Assign library networkx untuk membuat grafik relasi antar node
G = nx.Graph()

# Tambahkan node dari baris 'Node name' pada file nodes.xlsx ke variabel S 
for index, row in nodes_df.iterrows():
    G.add_node(row['Node name'])
# Tambahkan edge dari baris 'Source ID' dan 'Target ID' pada file edges.xlsx ke variabel S
for index, row in edges_df.iterrows():
    G.add_edge(row['Source ID'], row['Target ID'])

# Gambarkan grafik relasi pada variabel G
nx.draw(G,with_labels=True, font_size=9, node_color="lightblue", edge_color="red")
```

**Tampilan Jaringan Semantik Hasil Searh Filtering**

> Dalam melakukan filter pencarian, dibutuhkan variabel baru untuk mendapat inputan dari user.
Kemudian ambil setiap item dari baris nama node, source, atau target yang sesuai dengan inputan user.


```python
i = input("Search : ")
node_search = nodes_df[nodes_df['Node name'] == i]
edge_search = edges_df[(edges_df['Source ID'] == i) | (edges_df['Target ID'] == i)]
```

```python
# Assign library networkx utnuk membuat grafik relasi antar node
S = nx.Graph()

# Tambahkan node dari baris 'Node name' pada file nodes.xlsx ke variabel S 
for index, row in node_search.iterrows():
    S.add_node(row['Node name'])
# Tambahkan edge dari baris 'Source ID' dan 'Target ID' pada file edges.xlsx ke variabel S
for index, row in edge_search.iterrows():
    S.add_edge(row['Source ID'], row['Target ID'])

# Gambarkan grafik relasi pada variabel S
nx.draw(S,with_labels=True, node_color="lightblue", edge_color="red")
```
